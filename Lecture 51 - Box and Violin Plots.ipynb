```python
# The normal imports
import numpy as np
from numpy.random import randn
import pandas as pd

# Import the stats library from numpy
from scipy import stats

# These are the plotting modules adn libraries we'll use:
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# Command so that plots appear in the iPython Notebook
%matplotlib inline
```

```python
data1 = randn(100)
data2 = randn(100)
```

```python
sns.boxplot([data1,data2])
```

```python
sns.boxplot([data1,data2],whis=np.inf)  # To include all the outliers
```

```python
sns.boxplot(data1, vert = False)  #vertical boxplot
```

```python
# Normal Distribution
data1 = stats.norm(0,5).rvs(100)

# Two gamma distributions. Concatenated together
data2 = np.concatenate([stats.gamma(5).rvs(50)-1,-1*stats.gamma(5).rvs(50)])

# Box Plot both data1 and data2

sns.boxplot([data1,data2], whis =np.inf)
```

```python
sns.boxplot([data1,data2], whis =np.inf, vert = False)
```

```python
sns.violinplot([data1,data2])
```

```python
sns.violinplot(data2,bw=0.01)  #bw= bandwidth
```

```python
sns.violinplot(data1, inner = 'stick')
```
